```
import PyPDF2

input_pdf_path = "MPESA_Statement_2023-04-06_to_2025-04-06_2547xxxxxx956.pdf"
pdf_password = "331955"

reader = PyPDF2.PdfReader(input_pdf_path)
if reader.is_encrypted:
    reader.decrypt(pdf_password)
    print("PDF decrypted successfully.")
else:
    print("PDF is not encrypted.")

writer = PyPDF2.PdfWriter()
for page in reader.pages:
    writer.add_page(page)

unlocked_pdf_path = "unlocked_statement.pdf"
with open(unlocked_pdf_path, "wb") as f:
    writer.write(f)
```

```
import pdfplumber
import pandas as pd

all_rows = []

with pdfplumber.open("unlocked_statement.pdf") as pdf:
    for i, page in enumerate(pdf.pages):
        tables = page.extract_tables()
        for table in tables:
            if table and len(table[0]) >= 5:  # Valid table
                headers = table[0]
                for row in table[1:]:
                    if len(row) == len(headers):
                        all_rows.append(dict(zip(headers, row)))

df = pd.DataFrame(all_rows)
df.head()
print(f"\n✅ Total rows extracted: {len(df)}")
```
